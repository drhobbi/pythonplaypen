## Working with notebook cells

A notebook is a document consisting of a sequence of cells. A cell can either contain text or code. Generally the purpose of text cells is to provide instructions about what to do in the lab, although occasionally you will also need to edit text cells as part of an exercise (this will be made obvious in the lab). In contrast, code cells contain Python code which can be executed—most of the exercises involve writing and modifying code cells. If a code cell produces output, that output will be shown directly below the cell after it is run.

This is a text cell. I can display text, like the quote below:

Double-click on this text cell to enable editing, and then alter the quote to fix the spelling mistake.

I am another text cell. But my neighbour below is not, it's a code cell. Notice its different appearance with a grey background?

Select the code cell below by clicking on it once, and then press `<CTRL>+<ENTER>` (or click the "Play" button) to run the code inside and see what the output is.

```
for i in range(5, 0, -1):
    print(i)
print('Blast off!')
```

Edit the code cell above so that the countdown starts at 10 (just replace the `5` in the code with `10`). Once you're done, run the cell again to see the modified version of the code in action.

Save your changes to the notebook by clicking File > Save, or using the shortcut `<CTRL>+S`.

## Customising Google Colab settings

It is common practice in the Python programming community to use 4 spaces for indentation of code. To make it easier to follow this convention, we can customise the settings in Google Colab so that pressing the Tab key inserts 4 spaces for us.

In the menu at the top of the page, click "Tools" > "Settings", then go to the "Editor" tab. Find the "Indentation width in spaces" option and set it to 4. Click "Save" and Google Colab will remember this setting for all of the labs.
